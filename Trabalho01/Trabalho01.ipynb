[Biblioteca python apyori](https://pypi.org/project/apyori/) - https://pypi.org/project/apyori/

```python
import pandas as pd 
import numpy as np 
from apyori import apriori
```

```python
# importar os dados e converter em lista 
transacoes = []
with open('../MO433/Trabalho01/dados/retail.dat', 'r') as file:
    for l in file.readlines():
        #print(l.split())
        transacoes.append(l.split())
```

```python
# remover cabeçalho (primeira linha)
transacoes.pop(0)
```

```python
# suporte mínimo de 0.005 e confiança minima de 0.9
regras = apriori(transacoes, min_support = 0.005, min_confidence = 0.9)
```

```python
# resultado e quantidade de regras geradas 
resultados = list(regras)
len(resultados)
```

```python
# Conjunto de regras geradas
resultados
```

```python
items_base = []
items_add = []
suporte = []
confianca = []
lift = []

for resultado in resultados:
    #print(resultado)
    supor = resultado[1]
    #print(resultado_posicao)
    resultadoRegras = resultado[2]
    #print(resultadoRegras)
    for resultadoRegra in resultadoRegras:

        base = list(resultadoRegra[0])
        adicional  = list(resultadoRegra[1])
        conf = resultadoRegra[2]
        l = resultadoRegra[3]

        items_base.append(base)
        items_add.append(adicional)
        suporte.append(supor)
        confianca.append(conf)
        lift.append(l)    
```

```python
df_Regras = pd.DataFrame({'items_base': items_base, 'items_add': items_add, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
```

```python
# Visulalização das regras com suporte, confiança e lift
# Foram encontardas 37 regras e pode ser observado que o item 38 é encontrado na maioria das regras geradas
df_Regras
```
