### Detecção de outliers

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
from sklearn import svm
from sklearn.ensemble import IsolationForest
from functools import reduce
```

```python
df = pd.read_csv('../Trabalho03/dados/dados3.csv')
```

```python
df.head()
```

### Local Outlier Factor (desnsidade)
Os testes com o valor do n_neighbors apresentou pequenas variações no resultado.

O valor da contamination = 0.01 (1%) obiteve 10 anomalias (outliers).  

```python
clf = LocalOutlierFactor(n_neighbors=100, contamination = 0.01)
resultLof = clf.fit_predict(df)
```

```python
lof_outliers = []
for index, i in np.ndenumerate(resultLof):
    if(i == -1):
       lof_outliers.append(index[0])
lof_outliers = np.array(lof_outliers)
print("Número de anomalias (outliers): ", np.size(lof_outliers))
print("Index dos elementos com Anomalias (outliers): ", lof_outliers)
```

```python
# Dados com anomalias encontrados - Local Outlier Factor
df.loc[df.index.isin(lof_outliers)]
```

### One-class SVM

O valor da contamination = 0.01 (1%) obiteve 7 anomalias (outliers).  

```python
clf_svm = svm.OneClassSVM(nu=0.01, kernel="rbf").fit(df)
one_class_svm =  clf_svm.predict(df)
```

```python
svm_outlier = []
for index, i in np.ndenumerate(one_class_svm):
    if(i == -1):
      svm_outlier.append(index[0])
svm_outlier = np.array(svm_outlier)
print("Número de anomalias (outliers): ", np.size(svm_outlier))
print("Index dos elementos com Anomalias (outliers): ", svm_outlier)
```

```python
# Dados com anomalias encontrados - svm
df.loc[df.index.isin(svm_outlier)]
```

### IsolationForest

O valor da contamination = 0.01 (1%) obiteve 10 anomalias (outliers).  

```python
clf_IsolationForest = IsolationForest(n_estimators=100, contamination = 0.01, random_state=42).fit(df)
IsolationForest_result = clf_IsolationForest.predict(df)
```

```python
IsolationForest_outlier = []
for index, i in np.ndenumerate(IsolationForest_result):
    if(i == -1):
      IsolationForest_outlier.append(index[0])
IsolationForest_outlier = np.array(IsolationForest_outlier)
print("Número de anomalias (outliers): ", np.size(IsolationForest_outlier))
print("Index dos elementos com Anomalias (outliers): ", IsolationForest_outlier)
```

```python
# Dados com anomalias encontrados - Isolation Forest
df.loc[df.index.isin(IsolationForest_outlier)]
```

### Intersecção dos resultados de cada algoritmo (Combinação)

```python
resultado = reduce(np.intersect1d,(lof_outliers, svm_outlier, IsolationForest_outlier))
df.loc[df.index.isin(resultado)]
```
